```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mvc.data import transformation
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 3000)
```

```python
y = np.stack([3*np.sin(x) + np.exp(np.cos(x)) + np.random.randn(3000)*0.5, np.exp(np.sin(x)) + 2*np.cos(x) + np.random.randn(3000)*0.5])
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0], linewidth=0.5, label='ch1')
plt.plot(x, y[1], linewidth=0.5, color='red', label='ch2')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.Jittering()(y)[0], linewidth=0.5, label='Jittering')
plt.title('Jittering')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.HorizontalFlipping(randomize=False)(y)[0], linewidth=0.5, label='HorizontalFlipping')
plt.title('HorizontalFlipping')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.VerticalFlipping(randomize=False)(y)[0], linewidth=0.5, label='VerticalFlipping')
plt.title('VerticalFlipping')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.MagnitudeWarping(sigma=0.5)(y)[0], linewidth=0.5, label='MagnitudeWarping')
plt.title('MagnitudeWarping')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.TimeWarping(sigma=0.5)(y)[0], linewidth=0.5, label='TimeWarping')
plt.title('TimeWarping')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.Scaling(direction='both')(y)[0], linewidth=0.5, label='Scaling')
plt.title('Scaling')
plt.legend()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x, y[0]+5, linewidth=0.5, label='Ori')
plt.plot(x, transformation.RandomCropping(size=2000, fill=0, padding_mode='edge')(y)[0], linewidth=0.5, label='RandomCropping')
plt.title('RandomCropping')
plt.legend()
```
